```python
#Opening a file with block helps us that we dont have to close the file

car_specs = []

# file close ince you come out of with block
with open('ibm_stock_price-Copy1.csv','r') as file_again:
    for car in file_again:
        car_specs.append(car.rstrip("\n"))#Returns a copy of the string with trailing characters removed.
```
